```python
from tkinter import *
from tkinter import ttk,filedialog,messagebox
import tkinter as tk
from deep_translator import GoogleTranslator

class TranslatorGUI:

    def __init__(self, master):
        self.master = master
        master.title("Language Translator")
        
        self.source_label = Label(master, text="Source Language:")
        self.source_label.grid(row=0, column=0, padx=10, pady=10)

        self.target_label = Label(master, text="Target Language:")
        self.target_label.grid(row=1, column=0, padx=10, pady=10)

        self.source_var = StringVar()
        self.source_dropdown = ttk.Combobox(master, textvariable=self.source_var)
        self.source_dropdown['values'] =dis

        self.source_dropdown.current(0)
        self.source_dropdown.grid(row=0, column=1, padx=10, pady=10)

        self.target_var = StringVar()
        self.target_dropdown = ttk.Combobox(master, textvariable=self.target_var)
        self.target_dropdown['values'] =dis
        self.target_dropdown.current(1)
        self.target_dropdown.grid(row=1, column=1, padx=10, pady=10)

        self.text_label = Label(master, text="Enter text to translate:")
        self.text_label.grid(row=2, column=0, padx=10, pady=10)

        self.text_box = Text(master, height=5, width=30)
        self.text_box.grid(row=2, column=1, padx=10, pady=10)

        self.translate_button = Button(master, text="Translate", command=self.translate)
        self.translate_button.grid(row=3, column=1, padx=10, pady=10)

        
        self.translation_label = Label(master, text="")
        self.translation_label.grid(row=4,column=1,padx=20, pady=20)
        
        self.so = StringVar()
        self.so_dropdown = ttk.Combobox(master, textvariable=self.so)
        self.so_dropdown['values'] =dis
        
        self.so_dropdown.current(0)
        self.so_dropdown.grid(row=5,column=0, padx=10, pady=10)
        
        self.va = StringVar()
        self.va_dropdown = ttk.Combobox(master, textvariable=self.va)
        self.va_dropdown['values'] =dis
        
        self.va_dropdown.current(0)
        self.va_dropdown.grid(row=5,column=1, padx=10, pady=10)
        
        self.Name = Label(master,text="Output Name:")
        self.Name.grid(row=6,column=0, padx=10, pady=10)

        self.Nentry = Entry(master)
        self.Nentry.grid(row=6,column=1, padx=10, pady=10)

        self.import_button = tk.Button(master, text="Import File", command=self.ap)
        self.import_button.grid(row=7, column=0, padx=10, pady=10)

        self.generate_button = tk.Button(master, text="Generate Output", command=self.generate_output)
        self.generate_button.grid(row=7, column=1, padx=10, pady=10)
    
    def ap(self):
        global file_data
        global file_path
        global translated
        file_path = filedialog.askopenfilename(filetypes=(("Text files", "*.txt"),))
        if file_path:
            with open(file_path, "r",encoding='utf-8') as f:
                file_data = f.read()
                #print(file_data)
                #print(fl[f'{self.so.get()}'],fl[f'{self.va.get()}'])
                translated = GoogleTranslator(source=fl[f'{self.so.get()}'], target=fl[f'{self.va.get()}']).translate(file_data)
                #print(translated)
    def generate_output(self):
        try: 
            if file_path: 
                # Translate the text to the target language

                with open(f'{self.Nentry.get()}.txt', 'w', encoding='utf-8') as f:
                    f.write(translated)

                    print(f"Output file generated as {self.Nentry.get()}.txt")
        except NameError:
            #print("No file imported or translated yet")
            messagebox.showerror('File do not exist', 'No file imported or translated yet')

    def translate(self):
        source_lang = fl[f'{self.source_var.get()}']
        target_lang = fl[f'{self.target_var.get()}']
        text = self.text_box.get("1.0", END).strip()

        if text == "":
            self.translation_label.config(text="Please enter some text to translate.")
        else:
            try:
                translated_text = GoogleTranslator(source=source_lang, target=target_lang).translate(text)
                self.translation_label.config(text=translated_text)
            except:
                #self.translation_label.config(text="Translation failed. Please try again.")
                self.translation_label.config(text=source_lang)

root = Tk()
translator_gui = TranslatorGUI(root)
root.mainloop()

```
